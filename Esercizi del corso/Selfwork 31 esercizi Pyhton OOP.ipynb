```python
# Esercizio 1
# Creare una classe Person , estenderla con due sottoclassi Dipendente e Studente. 
# Una classe deve essere astratta e ogni classe deve avere delle proprietà specializzanti
from abc import ABC, abstractmethod
from datetime import date

class Person:
    def __init__(self, nome, cognome, data_nascita, luogo_nascita):
        self.name = nome
        self.surname = cognome
        self.birth_day = data_nascita
        self.birth_place = luogo_nascita
    def calcola_eta(self):
        return date.today().year -self.birth_day.year
    def presentazione(self):
        print(f"Piacere, mi chiamo {self.name} {self.surname}, ho {self.calcola_eta()}, e sono nato a {self.birth_place}")

class Dipendente(Person, ABC):
    def __init__(self, nome, cognome, data_nascita, luogo_nascita, identificativo, stipendio):
        super().__init__(nome, cognome, data_nascita, luogo_nascita)
        self.identification = identificativo
        self.salary = stipendio
    @abstractmethod
    def lavora(self):
        pass

class Studente(Person):
    def __init__(self, nome, cognome, data_nascita, luogo_nascita, matricola, voto, materia):
        super().__init__(nome, cognome, data_nascita, luogo_nascita)
        self.grade = voto
        self.number = matricola
        self.subject = materia
    def saluta(self):
        print(f"Ciao! mi chiamo {self.name} {self.surname}, ho {self.calcola_eta()} anni, e studio {self.subject}")

s = Studente("Vittorio", "Fiorita", date(1995, 4, 24), "Roma", "MAT001", 28, "Coding IA")
s.saluta()
```

```python
#Esercizio 2:
#Estendere Dipendenti con 2 sottoclassi Docente e Freelance. 
# Una classe deve essere astratta e ogni classe deve avere delle proprietà specializzanti
class Docente(Dipendente):
    def __init__(self, nome, cognome, data_nascita, luogo_nascita, identificativo, stipendio, materia):
        super().__init__(nome, cognome, data_nascita, luogo_nascita, identificativo, stipendio)
        self.subject = materia
    def lavora(self):
        print(f"Buongiorno, sono il professor {self.surname} {self.name}, ho {self.calcola_eta()} anni ed insegno {self.subject}")

class Freelance(Dipendente):
    def __init__(self, nome, cognome, data_nascita, luogo_nascita, identificativo, stipendio, specializzazione):
        super().__init__(nome, cognome, data_nascita, luogo_nascita, identificativo, stipendio)
        self.specialization = specializzazione
    def lavora(self):
        print(f"Buongiorno, sono {self.surname} {self.name}, ho {self.calcola_eta()} anni e lavoro come {self.specialization}")

d = Docente("Mario", "Rossi", date(1980, 5, 10), "Milano", "ID001", 2000, "Matematica")
d.lavora()

f = Freelance("Luigi", "Bianchi", date(1990, 3, 15), "Napoli", "ID002", 1500, "Programmatore Web")
f.lavora()
```
