```python
#Esercizio 2 - selezioni

numero = int(input("Inserisci un numero: "))

if (numero >= 90 and numero <= 100):
    print("A")
elif (numero >= 80 and numero <= 89):
    print("B")
elif (numero >= 70 and numero <= 79):
    print("C")
elif (numero >= 60 and numero <= 69):
    print("D")
else:
    print("F")
```

```python
#Esercizio 3 - selezioni

numero_a = int(input("Inserisci un numero: "))
numero_b = int(input("Inserisci un numero: "))
operazione = input("Inserisci l'operazione (+, -, *, /): ")

if operazione == "+":
    print(numero_a + numero_b)
elif operazione == "-":
    print(numero_a - numero_b)
elif operazione == "*":
    print(numero_a * numero_b)
elif ((operazione == "/") and numero_b == 0):
    print("Non puoi dividere per zero, CAPRA!")
elif operazione == "/":
    print(numero_a / numero_b)
else:
    print("l'operazione scelta non è valida")
```

```python
#Esercizio 4 - selezioni

anno = int(input("Inserisci un anno: "))

if (anno % 400 == 0):
    print("l'anno", anno, "è bisestile")
elif (anno % 100 == 0):
    print("l'anno", anno, "non è bisestile")
elif (anno % 4 == 0):
    print("l'anno", anno, "è bisestile")
else:
    print("l'anno", anno, "non è bisestile")
```
